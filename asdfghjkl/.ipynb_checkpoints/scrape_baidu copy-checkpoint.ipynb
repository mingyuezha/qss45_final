### Functions

Baidu fields from SerpAPI:
- `answer_box` — AI or health card content; analog to Google AI Overview
- `knowledge_graph` — structured entity card for well-known conditions
- `organic_results` — standard ranked results
- `related_searches` — Baidu's equivalent of People Also Ask

**Confirmed `answer_box` type values** (from SearchAPI/SerpAPI docs):
- `ai_search` — ERNIE AI answer; includes a `references` array with cited sources ✅
- `ai_content` — simpler AI answer; no references array
- `dictionary_result` — translation box (not health-relevant)
- `finance_results`, `line_chart_result` — non-health types

The `references` array on `ai_search` responses is the citation goldmine —
directly comparable to Google AI Overview's cited sources.

```python
def query_baidu(search_query: str, api_key: str) -> dict:
    """Query Baidu via SerpAPI. Queries should be in Mandarin."""
    client = serpapi.Client(api_key=api_key)
    results = client.search({
        "engine": "baidu",
        "q": search_query,
    })
    return results
```

```python
def parse_answer_box(response: dict) -> dict | None:
    
    box = response.get("ai_overview")
    if not box:
        return None

    # Extract references — only present on ai_search type
    references = [
        {
            "title":   r.get("title", ""),
            "snippet": r.get("snippet", ""),
            "source":  r.get("source", ""),
        }
        for r in box.get("references", [])
    ]

    return {
        "title":      box.get("title", ""),
        "answer":     box.get("answer", ""),
        "snippet":    box.get("snippet", ""),
        "link":       box.get("link", ""),
        "source":     box.get("source", ""),
        "references": references,
        "raw":        dict(box),
    }
```

```python
def parse_organic_results(response: dict, top_n: int = 5) -> list:
    """Extract top organic results for source-type classification."""
    organics = response.get("organic_results", [])
    return [
        {
            "position": r.get("position", ""),
            "title":    r.get("title", ""),
            "link":     r.get("link", ""),
            "snippet":  r.get("snippet", ""),
            "source":   r.get("source", ""),
        }
        for r in organics[:top_n]
    ]
```

```python
def extract_related_searches(response: dict) -> list:
    """Extract Baidu related searches (analog to Google's People Also Ask)."""
    return [
        item.get("query", "")
        for item in response.get("related_searches", [])
    ]
```

### Pipeline

Captures per query:
1. **answer_box** — AI or health card content (analog to Google AI Overview)
   - `is_ai` flag identifies `ai_search` / `ai_content` types
   - `references` contains citations for `ai_search` responses
2. **knowledge_graph** — structured entity card
3. **organic_results** — top 5 for source classification
4. **related_searches** — seed queries for future rounds

Output columns mirror the Google notebook for easy merging on `query_en`.

```python
def run_baidu_pipeline(queries: list, api_key: str) -> pd.DataFrame:
    rows = []
    total = len(queries)

    for i, (english_label, mandarin_query) in enumerate(queries):
        print(f"[{i+1}/{total}] EN: '{english_label}' | ZH: '{mandarin_query}'")

        try:
            response        = query_baidu(mandarin_query, api_key)
            answer_box      = parse_answer_box(response)
            organics        = parse_organic_results(response, top_n=5)
            related         = extract_related_searches(response)

            rows.append({
                # Identifiers — merge key with Google dataset is query_en
                "query_en":              english_label,
                "query_zh":              mandarin_query,

                # Answer box (AI overview analog)
                "has_answer_box":        answer_box is not None,
                "answer_box_title":      answer_box["title"]      if answer_box else None,
                "answer_box_answer":     answer_box["answer"]     if answer_box else None,
                "answer_box_snippet":    answer_box["snippet"]    if answer_box else None,
                "answer_box_link":       answer_box["link"]       if answer_box else None,
                "answer_box_source":     answer_box["source"]     if answer_box else None,
                "answer_box_references": answer_box["references"] if answer_box else None,
                "answer_box_raw":        answer_box["raw"]        if answer_box else None,

                # Organic + related
                "organic_results":       organics,
                "related_searches":      related,
            })

        except Exception as e:
            print(f"   ERROR on '{mandarin_query}': {e}")
            rows.append({
                "query_en":              english_label,
                "query_zh":              mandarin_query,
                "answer_box_title":      None,
                "answer_box_answer":     None,
                "answer_box_snippet":    None,
                "answer_box_link":       None,
                "answer_box_source":     None,
                "answer_box_references": None,
                "answer_box_raw":        None,
                "organic_results":       None,
                "related_searches":      None,
            })

        time.sleep(1.5)

    df = pd.DataFrame(rows)
    print(f"\nTotal rows: {len(df)}")
    print(f"Queries with answer_box:      {df['has_answer_box'].sum()}")
    print(f"Queries without answer_box:   {(~df['has_answer_box']).sum()}")
    return df
```

```python
df = run_baidu_pipeline(queries, api_key)
df
```

### Inspection helpers

```python
# Preview answer_box content for a topic section
topic_filter = "cancer"

subset = df[df["query_en"].str.contains(topic_filter)]
for _, row in subset.iterrows():
    print(f"\n=== {row['query_en']} ({row['query_zh']}) ===")
    print(f"  answer_box:      {row['has_answer_box']}")
    print(f"  knowledge_graph: {row['has_knowledge_graph']}")
    if row["has_answer_box"]:
        print(f"  type:    {row['answer_box_type']}")
        print(f"  answer:  {str(row['answer_box_answer'])[:300]}")
        refs = row["answer_box_references"]
        if refs:
            print(f"  references ({len(refs)}):")
            for r in refs:
                print(f"    - {r['source']}: {r['title']}")
```

```python
# Suppression check — no answer_box AND no knowledge_graph
# Absence on legally sensitive topics is a finding, not a scraping error
suppressed = df[(~df["has_answer_box"]) & (~df["has_knowledge_graph"])]
print(f"Fully suppressed queries ({len(suppressed)}):")
print(suppressed[["query_en", "query_zh"]].to_string(index=False))

print("\n--- Answer box type breakdown ---")
print(df["answer_box_type"].value_counts(dropna=False))

print("\n--- AI vs non-AI answer boxes ---")
print(df["answer_box_is_ai"].value_counts(dropna=False))
```
