```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import numpy as np
import re

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams["figure.dpi"] = 120
```

```python
combined = pd.read_csv("data/df.csv")
```

```python
# ── reference parsing helpers ─────────────────────────────────────────────────
_INSTITUTIONAL  = re.compile(
    r'(\.gov|\.edu|who\.int|nih\.gov|cdc\.gov|nhs\.uk|'
    r'ncbi\.nlm\.nih|pubmed|un\.org|nhc\.gov\.cn|moh\.gov\.cn)', re.I)
_ACADEMIC       = re.compile(
    r'(pubmed|ncbi|nature\.com|nejm\.org|thelancet|'
    r'bmj\.com|jamanetwork|springer|wiley|tandfonline|'
    r'sciencedirect|academic\.oup|cnki)', re.I)
_HEALTH_PLAT    = re.compile(
    r'(webmd|mayoclinic|healthline|medicinenet|rxlist|bohe|youlai|'
    r'drugs\.com|medscape|clevelandclinic|hopkinsmedicine|'
    r'cancer\.org|heart\.org|diabetes\.org)', re.I)
_ENCYCLOPEDIA   = re.compile(r'(wikipedia|britannica|baike\.baidu)', re.I)
_CHINESE_TLD    = re.compile(r'\.(cn|com\.cn)(/|$)')
_CHINESE_DOM    = re.compile(
    r'(baidu|zhihu|sina|sohu|163\.com|qq\.com|weibo|bilibili|'
    r'ifeng|tencent|alibaba|taobao|jd\.com|xinhua|'
    r'people\.com|chinadaily|cctv|gov\.cn|nhc\.gov|moh\.gov)')

def _classify_source(url):
    if not isinstance(url, str): return "unknown"
    if _INSTITUTIONAL.search(url):  return "institutional"
    if _ACADEMIC.search(url):       return "academic"
    if _HEALTH_PLAT.search(url):    return "health_platform"
    if _ENCYCLOPEDIA.search(url):   return "encyclopedia"
    return "lay"

def _classify_locality(url):
    if not isinstance(url, str): return "unknown"
    if _CHINESE_TLD.search(url) or _CHINESE_DOM.search(url): return "chinese"
    return "western"

def _domain(url):
    if not isinstance(url, str): return "unknown"
    m = re.search(r'https?://([^/]+)', url)
    return m.group(1).removeprefix("www.") if m else "unknown"

def _parse_refs(row):
    raw = row["all_references"]
    if not isinstance(raw, str) or raw.strip() in ("", "[]"):
        return []
    try:
        items = ast.literal_eval(raw)
    except Exception:
        return []
    if not isinstance(items, list):
        return []
    out = []
    for r in items:
        if not isinstance(r, dict): continue
        url = r.get("link", "")
        out.append({
            "platform":    row["platform"],
            "section":     row["section"],
            "topic":       row["topic"],
            "title":       r.get("title", ""),
            "url":         url,
            "domain":      _domain(url),
            "source_type": _classify_source(url),
            "locality":    _classify_locality(url),
        })
    return out

# ── build all_refs and per-query count columns ────────────────────────────────
all_rows, count_rows = [], []
for _, row in combined.iterrows():
    refs = _parse_refs(row)
    all_rows.extend(refs)
    count_rows.append({
        "num_refs":          len(refs),
        "num_institutional": sum(1 for r in refs if r["source_type"] == "institutional"),
        "num_lay":           sum(1 for r in refs if r["source_type"] == "lay"),
        "num_western":       sum(1 for r in refs if r["locality"] == "western"),
        "num_chinese":       sum(1 for r in refs if r["locality"] == "chinese"),
    })

all_refs = pd.DataFrame(all_rows)
combined = pd.concat([combined.reset_index(drop=True),
                      pd.DataFrame(count_rows)], axis=1)

print(f"all_refs: {len(all_refs)} reference rows")
print(f"combined columns: {list(combined.columns)}")
```

## 1 — AI Overview Return Rate

What proportion of queries return an AI overview, by platform and topic section?

```python
SECTION_ORDER = ['common_conditions', 'controversial', 'legally_variable', 'tcm']
```

```python
rate = (
    combined
    .groupby(["platform", "section"])["has_ai_overview"]
    .mean()
    .reset_index()
    .rename(columns={"has_ai_overview": "ai_overview_rate"})
)
rate["section"] = pd.Categorical(rate["section"])
rate = rate.sort_values("section")

fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(data=rate, x="section", y="ai_overview_rate", hue="platform",
            palette={"google": "#4285F4", "baidu": "#E34234"}, ax=ax)
ax.set_title("AI Overview Return Rate by Section", fontsize=13, fontweight="bold")
ax.set_xlabel("")
ax.set_ylabel("Proportion of queries with AI overview")
ax.yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1))
ax.set_xticklabels([s.replace("_", " ").title() for s in SECTION_ORDER])
plt.tight_layout()
plt.savefig("fig_ai_overview_rate.png")
plt.show()

rate
```

### Suppression table — queries with NO AI overview on either platform

```python
suppression = (
    combined
    .groupby(["query_en", "section", "topic", "platform"])["has_ai_overview"]
    .first()
    .unstack("platform")
    .reset_index()
)
suppression["suppressed_on"] = suppression.apply(
    lambda r: "both" if not r.get("google", True) and not r.get("baidu", True)
              else ("google only" if not r.get("google", True)
              else ("baidu only"  if not r.get("baidu", True)
              else "neither")), axis=1
)
suppression[suppression["suppressed_on"] != "neither"][["query_en", "topic", "suppressed_on"]]
```

## 2 — Citation Count

How many references does each platform provide per query, by section?

```python
# Compute means and standard errors
summary = (
    combined
    .groupby(["section", "platform"])["num_refs"]
    .agg(mean="mean", se="sem")
    .reset_index()
)

# Ensure section order
summary["section"] = pd.Categorical(
    summary["section"],
    categories=SECTION_ORDER,
    ordered=True
)

summary = summary.sort_values("section")

# Plot
fig, ax = plt.subplots(figsize=(6, 4))

palette = {"google": "#4285F4", "baidu": "#E34234"}

for platform in summary["platform"].unique():
    subset = summary[summary["platform"] == platform]

    ax.errorbar(
        x=subset["section"],
        y=subset["mean"],
        yerr=subset["se"],
        fmt="o",          # points only
        linestyle="none", # remove connecting lines
        capsize=4,
        markersize=7,
        label=platform.title(),
        color=palette[platform]
    )

ax.set_title(
    "Average Number of References per Query",
    fontsize=13,
    fontweight="bold"
)

ax.set_xlabel("")
ax.set_ylabel("Reference count")

ax.set_xticks(range(len(SECTION_ORDER)))
ax.set_xticklabels(
    [s.replace("_", " ").title() for s in SECTION_ORDER]
)

ax.legend(title="Platform")

plt.tight_layout()
plt.savefig("fig_ref_scatter_errorbars.png", dpi=300, bbox_inches="tight")
plt.show()
```

## 3 — Source Type Distribution

What types of sources does each platform cite?

```python
SOURCE_TYPES  = ["institutional", "academic", "health_platform", "encyclopedia", "lay", "unknown"]
SOURCE_COLORS = ["#2196F3", "#4CAF50", "#FF9800", "#9C27B0", "#F44336", "#9E9E9E"]

type_dist = (
    all_refs
    .groupby(["platform", "source_type"])
    .size()
    .reset_index(name="count")
)
type_totals = type_dist.groupby("platform")["count"].transform("sum")
type_dist["pct"] = type_dist["count"] / type_totals

pivot = type_dist.pivot(index="platform", columns="source_type", values="pct").fillna(0)
pivot = pivot.reindex(columns=[c for c in SOURCE_TYPES if c in pivot.columns])

pivot.plot(kind="bar", stacked=True, figsize=(8, 5),
           color=SOURCE_COLORS[:len(pivot.columns)])
plt.title("Source Type Distribution by Platform", fontsize=13, fontweight="bold")
plt.xlabel("")
plt.ylabel("Proportion of all citations")
plt.xticks(rotation=0)
plt.gca().yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1))
plt.legend(title="Source type", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.savefig("fig_source_type_dist.png")
plt.show()

pivot.round(3)
```

---
## 4 — Local vs Western Citation Ratio

Core hypothesis test: does Baidu cite Chinese sources more than Google?

```python
locality_dist = (
    all_refs
    .groupby(["platform", "locality"])
    .size()
    .reset_index(name="count")
)
loc_totals = locality_dist.groupby("platform")["count"].transform("sum")
locality_dist["pct"] = locality_dist["count"] / loc_totals

loc_pivot = locality_dist.pivot(index="platform", columns="locality", values="pct").fillna(0)

loc_pivot.plot(kind="bar", figsize=(7, 4),
               color=["#E34234", "#9E9E9E", "#4285F4", "#BDBDBD"])
plt.title("Citation Locality by Platform", fontsize=13, fontweight="bold")
plt.xlabel("")
plt.ylabel("Proportion of citations")
plt.xticks(rotation=0)
plt.gca().yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1))
plt.legend(title="Locality", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.savefig("fig_locality.png")
plt.show()

loc_pivot.round(3)
```

### Locality by topic section (heatmap)

```python
TOPIC_ORDER = combined["topic"].dropna().unique().tolist()
TOPIC_ORDER
```

```python
for locality_val in ["chinese", "western"]:
    heat = (
        all_refs[all_refs["locality"] == locality_val]
        .groupby(["platform", "topic"])
        .size()
        .div(
            all_refs.groupby(["platform", "topic"]).size()
        )
        .reset_index(name="pct")
        .pivot(index="platform", columns="topic", values="pct")
        .fillna(0)
    )
    heat = heat.reindex(columns=[t for t in TOPIC_ORDER if t in heat.columns])

    fig, ax = plt.subplots(figsize=(14, 3))
    sns.heatmap(heat, annot=True, fmt=".0%", cmap="Blues" if locality_val == "western" else "Reds",
                linewidths=0.5, ax=ax, vmin=0, vmax=1)
    ax.set_title(f"Proportion of '{locality_val}' citations by topic",
                 fontsize=12, fontweight="bold")
    ax.set_xlabel("")
    plt.tight_layout()
    plt.savefig(f"fig_heatmap_{locality_val}.png")
    plt.show()
```

## 5 — Top Cited Domains per Platform

```python
for platform in ["google", "baidu"]:
    top = (
        all_refs[all_refs["platform"] == platform]["domain"]
        .value_counts()
        .head(15)
    )
    fig, ax = plt.subplots(figsize=(8, 4))
    top.plot(kind="barh", ax=ax,
             color="#4285F4" if platform == "google" else "#E34234")
    ax.invert_yaxis()
    ax.set_title(f"Top 15 cited domains — {platform.title()}",
                 fontsize=12, fontweight="bold")
    ax.set_xlabel("Citation count")
    plt.tight_layout()
    plt.savefig(f"fig_top_domains_{platform}.png")
    plt.show()
```
