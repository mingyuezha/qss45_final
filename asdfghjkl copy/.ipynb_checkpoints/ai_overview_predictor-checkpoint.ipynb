# Predicting AI Overview Likelihood with CatBoost + SHAP
### Google Health Search · `data/all_google.csv`

**Goal:** Train a gradient-boosted tree model to predict whether a given health query returns an AI Overview, then use SHAP to explain *which features drive that prediction* and *how*.

**Features used:**
| Feature | Type | Description |
|---|---|---|
| `section` | Categorical | Health category (common conditions, controversial, legally variable, TCM) |
| `topic` | Categorical | Specific health topic (cancer, depression, HIV, …) |
| `location` | Categorical | Country where the search was performed |
| `first_word` | Categorical | First word of the query — captures phrasing style |
| `word_count` | Numeric | Number of words in the query |
| `char_count` | Numeric | Number of characters in the query |
| `is_question` | Binary | Does the query start with a question word? |
| `has_verb` | Binary | Does the query contain an action/state verb? |

```python
import re
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import shap
from catboost import CatBoostClassifier, Pool
from sklearn.metrics import (roc_auc_score, average_precision_score,
                             ConfusionMatrixDisplay, classification_report)
from sklearn.model_selection import StratifiedKFold

warnings.filterwarnings("ignore")
try:
    shap.initjs()
except Exception:
    pass  # only works inside a live Jupyter session

plt.rcParams.update({
    "figure.facecolor": "white",
    "axes.facecolor": "#f8fafc",
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.grid": True,
    "grid.color": "#e2e8f0",
    "grid.linewidth": 0.8,
    "font.family": "sans-serif",
    "axes.titlesize": 12,
    "axes.titleweight": "bold",
})

SEED = 42
print("All imports OK.")
```

## 1 · Load & Engineer Features

```python
df = pd.read_csv("data/all_google.csv")

# ── Target — column is already bool in this CSV ───────────────────────────────
df["target"] = df["has_ai_overview"].astype(int)

# ── Categorical features ──────────────────────────────────────────────────────
QUESTION_WORDS = (
    "what", "how", "why", "when", "where", "who", "which",
    "is", "are", "can", "do", "does", "will", "should"
)
VERBS = re.compile(
    r"\b(is|are|can|do|does|will|help|cause|treat|cure|prevent|"
    r"get|have|feel|make|work|use|take|manage|reduce|increase|stop)\b"
)

df["query_lower"] = df["query"].str.lower().str.strip()

df["section"]    = df["section"].fillna("").replace("", "unknown")
df["topic"]      = df["topic"].fillna("").replace("", "unknown")
df["location"]   = df["location"].str.strip()
df["first_word"] = df["query_lower"].str.split().str[0].fillna("other")

# ── Numeric features ──────────────────────────────────────────────────────────
df["word_count"]  = df["query_lower"].str.split().str.len()
df["char_count"]  = df["query_lower"].str.len()
df["is_question"] = df["query_lower"].str.startswith(QUESTION_WORDS).astype(int)
df["has_verb"]    = df["query_lower"].str.contains(VERBS).astype(int)

FEATURES = ["section", "topic", "location", "first_word",
            "word_count", "char_count", "is_question", "has_verb"]

CAT_FEATURES = ["section", "topic", "location", "first_word"]
NUM_FEATURES = ["word_count", "char_count", "is_question", "has_verb"]

X = df[FEATURES].copy()
y = df["target"].copy()

print(f"Dataset: {len(X)} rows  |  Target: {y.mean()*100:.1f}% positive")
print(f"\nClass counts:\n{y.value_counts().rename({0:'No AI Overview', 1:'AI Overview'})}")
X.head()
```

## 2 · Baseline Rate by Feature (EDA)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4.5))
fig.suptitle("AI Overview Rate by Key Categorical Features", fontsize=13, fontweight="bold")

PALETTE = {
    "section":  "#1565c0",
    "topic":    "#2e7d32",
    "location": "#6a1b9a",
}

SECTION_NAMES = {
    "common_conditions": "Common Conditions",
    "controversial":     "Controversial",
    "legally_variable":  "Legally Variable",
    "tcm":               "TCM",
    "unknown":           "Unknown",
}

TOPIC_NAMES = {
    "cancer": "Cancer", "depression": "Depression", "diabetes": "Diabetes",
    "hiv": "HIV", "hypertension": "Hypertension", "long_covid": "Long COVID",
    "abortion": "Abortion", "gender_affirming": "Gender Affirming",
    "hantavirus": "Hantavirus", "acupuncture": "Acupuncture",
    "cupping": "Cupping", "herbal": "Herbal", "unknown": "Unknown",
}

for ax, (col, name_map, color) in zip(axes, [
    ("section",  SECTION_NAMES,  "#1565c0"),
    ("topic",    TOPIC_NAMES,    "#2e7d32"),
    ("location", {},             "#6a1b9a"),
]):
    rates = df.groupby(col)["target"].mean().sort_values(ascending=True)
    labels = [name_map.get(k, k) for k in rates.index]
    bars = ax.barh(labels, rates.values * 100, color=color, alpha=0.80,
                   height=0.55, zorder=3)
    for bar, v in zip(bars, rates.values * 100):
        ax.text(v + 0.8, bar.get_y() + bar.get_height() / 2,
                f"{v:.1f}%", va="center", fontsize=9)
    ax.axvline(y.mean() * 100, color="#e53935", linewidth=1.4,
               linestyle="--", alpha=0.75, label=f"Overall: {y.mean()*100:.1f}%")
    ax.set_xlabel("AI Overview Rate (%)")
    ax.set_title(col.replace("_", " ").title())
    ax.set_xlim(0, 115)
    ax.legend(fontsize=8.5)
    ax.spines["left"].set_visible(True)
    ax.spines["left"].set_color("#cbd5e1")

plt.tight_layout()
plt.savefig("eda_rates.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 3 · 5-Fold Cross-Validation with CatBoost

```python
cat_feature_indices = [FEATURES.index(c) for c in CAT_FEATURES]

MODEL_PARAMS = dict(
    iterations=500,
    depth=4,
    learning_rate=0.05,
    l2_leaf_reg=5,
    random_seed=SEED,
    eval_metric="AUC",
    loss_function="Logloss",
    verbose=0,
    early_stopping_rounds=50,
)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

oof_preds  = np.zeros(len(y))
oof_proba  = np.zeros(len(y))
fold_aucs  = []
fold_aps   = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y), 1):
    X_tr, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]

    pool_tr  = Pool(X_tr, y_tr,  cat_features=CAT_FEATURES)
    pool_val = Pool(X_val, y_val, cat_features=CAT_FEATURES)

    model = CatBoostClassifier(**MODEL_PARAMS)
    model.fit(pool_tr, eval_set=pool_val)

    proba = model.predict_proba(pool_val)[:, 1]
    oof_proba[val_idx] = proba
    oof_preds[val_idx] = (proba >= 0.5).astype(int)

    auc = roc_auc_score(y_val, proba)
    ap  = average_precision_score(y_val, proba)
    fold_aucs.append(auc)
    fold_aps.append(ap)
    print(f"  Fold {fold}  |  AUC={auc:.4f}  |  AP={ap:.4f}  "
          f"|  best_iter={model.best_iteration_}")

print(f"\n{'─'*52}")
print(f"  OOF AUC  = {np.mean(fold_aucs):.4f} ± {np.std(fold_aucs):.4f}")
print(f"  OOF AP   = {np.mean(fold_aps):.4f} ± {np.std(fold_aps):.4f}")
print(f"  Baseline (always predict majority) AUC = 0.5000")
```

```python
from sklearn.metrics import roc_curve, precision_recall_curve

fig, axes = plt.subplots(1, 3, figsize=(15, 4.5))

# ── ROC curve ─────────────────────────────────────────────────────────────────
fpr, tpr, _ = roc_curve(y, oof_proba)
auc_score = roc_auc_score(y, oof_proba)
axes[0].plot(fpr, tpr, color="#1565c0", lw=2, label=f"OOF AUC = {auc_score:.3f}")
axes[0].plot([0, 1], [0, 1], "--", color="#90a4ae", lw=1, label="Random")
axes[0].fill_between(fpr, tpr, alpha=0.08, color="#1565c0")
axes[0].set_xlabel("False Positive Rate")
axes[0].set_ylabel("True Positive Rate")
axes[0].set_title("ROC Curve (Out-of-Fold)")
axes[0].legend(fontsize=9)

# ── Precision-Recall curve ────────────────────────────────────────────────────
prec, rec, _ = precision_recall_curve(y, oof_proba)
ap_score = average_precision_score(y, oof_proba)
axes[1].plot(rec, prec, color="#2e7d32", lw=2, label=f"OOF AP = {ap_score:.3f}")
axes[1].axhline(y.mean(), color="#90a4ae", linestyle="--", lw=1,
                label=f"Baseline = {y.mean():.2f}")
axes[1].fill_between(rec, prec, alpha=0.08, color="#2e7d32")
axes[1].set_xlabel("Recall")
axes[1].set_ylabel("Precision")
axes[1].set_title("Precision-Recall Curve (OOF)")
axes[1].legend(fontsize=9)

# ── Confusion matrix ──────────────────────────────────────────────────────────
ConfusionMatrixDisplay.from_predictions(
    y, oof_preds,
    display_labels=["No AI Overview", "AI Overview"],
    colorbar=False,
    ax=axes[2],
    cmap="Blues",
)
axes[2].set_title("Confusion Matrix (OOF, threshold=0.5)")

plt.tight_layout()
plt.savefig("cv_performance.png", dpi=150, bbox_inches="tight")
plt.show()

print("\nClassification Report (OOF):")
print(classification_report(y, oof_preds, target_names=["No AI Overview", "AI Overview"]))
```

## 4 · Full-Data Model for SHAP Analysis

SHAP values require a single trained model. We retrain on all 478 rows — using the median best iteration from CV to avoid overfitting.

```python
full_pool = Pool(X, y, cat_features=CAT_FEATURES)

full_model = CatBoostClassifier(
    **{**MODEL_PARAMS, "iterations": 300, "early_stopping_rounds": None}
)
full_model.fit(full_pool)

print(f"Full model trained  |  {full_model.tree_count_} trees")
print(f"Train AUC (in-sample): {roc_auc_score(y, full_model.predict_proba(full_pool)[:,1]):.4f}")
```

## 5 · CatBoost Feature Importance

```python
fi = pd.Series(
    full_model.get_feature_importance(full_pool),
    index=FEATURES
).sort_values(ascending=True)

FEAT_COLORS = {
    "section":    "#1565c0",
    "topic":      "#2e7d32",
    "location":   "#6a1b9a",
    "first_word": "#e65100",
    "word_count": "#00838f",
    "char_count": "#00838f",
    "is_question":"#f57f17",
    "has_verb":   "#f57f17",
}
colors = [FEAT_COLORS[f] for f in fi.index]

fig, ax = plt.subplots(figsize=(8, 4.5))
bars = ax.barh(fi.index, fi.values, color=colors, alpha=0.85, height=0.55, zorder=3)
for bar, v in zip(bars, fi.values):
    ax.text(v + 0.3, bar.get_y() + bar.get_height() / 2,
            f"{v:.1f}", va="center", fontsize=9.5)
ax.set_xlabel("Feature Importance (CatBoost default — PredictionValuesChange)")
ax.set_title("CatBoost Feature Importance")
ax.set_xlim(0, fi.max() * 1.2)
ax.spines["left"].set_visible(True)
ax.spines["left"].set_color("#cbd5e1")

legend_items = [
    mpatches.Patch(color="#1565c0", label="Categorical — query content"),
    mpatches.Patch(color="#e65100", label="Categorical — phrasing"),
    mpatches.Patch(color="#6a1b9a", label="Categorical — country"),
    mpatches.Patch(color="#00838f", label="Numeric — length"),
    mpatches.Patch(color="#f57f17", label="Binary — phrasing flag"),
]
ax.legend(handles=legend_items, fontsize=8.5, loc="lower right")

plt.tight_layout()
plt.savefig("feature_importance.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 6 · SHAP Analysis

```python
# CatBoost's native SHAP computation handles categoricals correctly
shap_matrix = full_model.get_feature_importance(full_pool, type="ShapValues")
# Last column is the bias term — drop it
shap_vals = shap_matrix[:, :-1]   # shape (n_samples, n_features)

shap_df = pd.DataFrame(shap_vals, columns=FEATURES)
print(f"SHAP matrix: {shap_vals.shape}")
print(f"\nMean |SHAP| per feature:")
print(shap_df.abs().mean().sort_values(ascending=False).round(4))
```

```python
# ── SHAP Summary (beeswarm) ───────────────────────────────────────────────────
# Build an Explanation object using the numeric version of X for display
X_display = X.copy()
for col in CAT_FEATURES:
    X_display[col] = X_display[col].astype(str)

explanation = shap.Explanation(
    values=shap_vals,
    base_values=np.full(len(shap_vals), shap_matrix[:, -1].mean()),
    data=X_display.values,
    feature_names=FEATURES,
)

fig, ax = plt.subplots(figsize=(9, 5))
shap.plots.beeswarm(explanation, max_display=8, show=False, color_bar=True)
plt.title("SHAP Beeswarm — Impact on AI Overview Prediction", fontsize=12, fontweight="bold", pad=10)
plt.tight_layout()
plt.savefig("shap_beeswarm.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# ── SHAP bar plot (mean |SHAP|) ───────────────────────────────────────────────
mean_shap = shap_df.abs().mean().sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.barh(mean_shap.index, mean_shap.values,
               color=[FEAT_COLORS[f] for f in mean_shap.index],
               alpha=0.85, height=0.55, zorder=3)
for bar, v in zip(bars, mean_shap.values):
    ax.text(v + 0.002, bar.get_y() + bar.get_height() / 2,
            f"{v:.4f}", va="center", fontsize=9)
ax.set_xlabel("Mean |SHAP value| (log-odds units)")
ax.set_title("Global SHAP Feature Importance")
ax.set_xlim(0, mean_shap.max() * 1.22)
ax.spines["left"].set_visible(True)
ax.spines["left"].set_color("#cbd5e1")
plt.tight_layout()
plt.savefig("shap_bar.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 7 · SHAP Deep-Dives: Per-Category Distributions

For each categorical feature, plot the distribution of SHAP values broken out by category value — showing which specific sections, topics, and countries push predictions up or down.

```python
def shap_by_category(feature, name_map=None, color="#1565c0", ax=None, title=None):
    tmp = df[[feature, "target"]].copy()
    tmp["shap"] = shap_df[feature].values
    means = tmp.groupby(feature)["shap"].mean().sort_values(ascending=True)

    labels = [name_map.get(k, k) if name_map else k for k in means.index]
    colors_bar = [color if v >= 0 else "#e53935" for v in means.values]

    if ax is None:
        fig, ax = plt.subplots(figsize=(7, 3.5))

    bars = ax.barh(labels, means.values, color=colors_bar, alpha=0.82,
                   height=0.55, zorder=3)
    ax.axvline(0, color="#546e7a", linewidth=1, linestyle="-", alpha=0.5)
    for bar, v in zip(bars, means.values):
        x_pos = v + 0.003 if v >= 0 else v - 0.003
        ha = "left" if v >= 0 else "right"
        ax.text(x_pos, bar.get_y() + bar.get_height() / 2,
                f"{v:+.3f}", va="center", fontsize=9, ha=ha)
    ax.set_xlabel("Mean SHAP value (log-odds)")
    ax.set_title(title or f"SHAP by {feature}")
    ax.spines["left"].set_visible(True)
    ax.spines["left"].set_color("#cbd5e1")
    lim = max(abs(means.values)) * 1.35
    ax.set_xlim(-lim, lim)
    return ax

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("Mean SHAP Value by Category (positive = more likely to return AI Overview)",
             fontsize=12, fontweight="bold")

shap_by_category("section",  SECTION_NAMES,  "#1565c0",
                 ax=axes[0], title="Section")
shap_by_category("topic",    TOPIC_NAMES,    "#2e7d32",
                 ax=axes[1], title="Topic")
shap_by_category("location", {},             "#6a1b9a",
                 ax=axes[2], title="Country")

plt.tight_layout()
plt.savefig("shap_by_category.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 8 · SHAP Heatmap: Section × Country Interaction

Does the model's response to `section` differ by country? We compute mean SHAP values for `section` within each country subgroup.

```python
tmp = df[["section", "location"]].copy()
tmp["shap_section"] = shap_df["section"].values
tmp["shap_location"] = shap_df["location"].values
# Combined SHAP = section + location contribution for each cell
tmp["shap_combined"] = tmp["shap_section"] + tmp["shap_location"]

sections_plot  = ["common_conditions", "controversial", "legally_variable", "tcm"]
countries_plot = ["Australia", "India", "United Kingdom", "United States"]

heat = np.zeros((len(countries_plot), len(sections_plot)))
for i, ctry in enumerate(countries_plot):
    for j, sec in enumerate(sections_plot):
        mask = (tmp["location"] == ctry) & (tmp["section"] == sec)
        heat[i, j] = tmp.loc[mask, "shap_combined"].mean() if mask.any() else np.nan

fig, ax = plt.subplots(figsize=(9, 4))
vmax = np.nanmax(np.abs(heat))
im = ax.imshow(heat, cmap="RdBu", aspect="auto", vmin=-vmax, vmax=vmax)

sec_labels = [SECTION_NAMES[s].replace("\n", " ") for s in sections_plot]
ax.set_xticks(range(len(sections_plot)))
ax.set_xticklabels(sec_labels, fontsize=11)
ax.set_yticks(range(len(countries_plot)))
ax.set_yticklabels(countries_plot, fontsize=11)

for i in range(len(countries_plot)):
    for j in range(len(sections_plot)):
        v = heat[i, j]
        if not np.isnan(v):
            tc = "white" if abs(v) > vmax * 0.55 else "#1e293b"
            ax.text(j, i, f"{v:+.2f}", ha="center", va="center",
                    fontsize=11, fontweight="bold", color=tc)

cbar = plt.colorbar(im, ax=ax, pad=0.02, shrink=0.85)
cbar.set_label("Combined SHAP (section + country)", fontsize=9)
ax.set_title("SHAP Interaction Heatmap: Section × Country\n"
             "(blue = pushes toward AI Overview, red = away)", fontsize=12, fontweight="bold")
ax.spines[:].set_visible(False)
ax.tick_params(length=0)
plt.tight_layout()
plt.savefig("shap_heatmap.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 9 · Predicted Probability Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))

# ── OOF probability distributions by true label ───────────────────────────────
for label, color, name in [(1, "#1565c0", "AI Overview"), (0, "#e53935", "No AI Overview")]:
    axes[0].hist(oof_proba[y == label], bins=20, alpha=0.55, color=color,
                 label=name, edgecolor="white", linewidth=0.5)
axes[0].axvline(0.5, color="#546e7a", linestyle="--", lw=1.5, label="Threshold 0.5")
axes[0].set_xlabel("Predicted Probability")
axes[0].set_ylabel("Count")
axes[0].set_title("OOF Predicted Probability by True Label")
axes[0].legend()

# ── Predicted probability by section (boxplot) ────────────────────────────────
df["oof_proba"] = oof_proba
sec_order = (df.groupby("section")["oof_proba"]
               .median()
               .sort_values(ascending=False)
               .index.tolist())
sec_data_box = [df.loc[df["section"] == s, "oof_proba"].values for s in sec_order]
sec_labels_box = [SECTION_NAMES.get(s, s) for s in sec_order]

bp = axes[1].boxplot(sec_data_box, labels=sec_labels_box, patch_artist=True,
                     medianprops=dict(color="white", linewidth=2),
                     whiskerprops=dict(color="#90a4ae"),
                     capprops=dict(color="#90a4ae"),
                     flierprops=dict(marker="o", markersize=4, alpha=0.5,
                                     markerfacecolor="#90a4ae", markeredgecolor="none"))
colors_box = ["#1565c0", "#1976d2", "#42a5f5", "#90caf9", "#e2e8f0"]
for patch, color in zip(bp["boxes"], colors_box):
    patch.set_facecolor(color)
    patch.set_alpha(0.75)
axes[1].axhline(0.5, color="#e53935", linestyle="--", lw=1.2, alpha=0.7, label="0.5 threshold")
axes[1].set_ylabel("Predicted Probability (OOF)")
axes[1].set_title("Predicted Probability by Section")
axes[1].legend(fontsize=9)
axes[1].set_ylim(0, 1.05)

plt.tight_layout()
plt.savefig("prob_distribution.png", dpi=150, bbox_inches="tight")
plt.show()
```
