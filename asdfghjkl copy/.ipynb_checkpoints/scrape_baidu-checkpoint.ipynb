### Functions

Baidu SerpAPI uses `engine: "baidu"` with `q` for the query.
- `ct=1` sets the content type to Chinese web results
- Queries should be in Mandarin for authentic results

Baidu results differ from Google structurally:
- No AI Overview block — Baidu's AI response (ERNIE) appears as an `ai_message` block
- `baidu_health` captures results from Baidu Health vertical (jiankang.baidu.com)
- `knowledge_graph` may appear for well-known medical conditions
- `organic_results` contains standard web results with title, link, snippet
- No direct equivalent to Google's PAA (related_questions) — Baidu has `related_searches` instead

```python
def query_baidu(search_query: str, api_key: str) -> dict:
    """
    Query Baidu via SerpAPI.
    
    Parameters
    ----------
    search_query : str
        The search term — ideally in Mandarin Chinese for authentic Baidu results.
    api_key : str
        SerpAPI key.

    Returns
    -------
    dict
        Raw SerpAPI response.
    """
    client = serpapi.Client(api_key=api_key)

    results = client.search({
        "engine": "baidu",
        "q": search_query,
    })

    return results
```

```python
def parse_baidu_ai_message(response: dict) -> dict:
    """
    Extract Baidu's ERNIE AI-generated answer (equivalent to Google AI Overview).
    
    Baidu's AI response appears under 'ai_message' in the SerpAPI response.
    It contains:
      - 'answer': the main AI-generated text
      - 'sources': list of cited sources (title, link, source name)
    
    Returns a normalized dict matching the structure used for Google AI Overviews.
    """
    ai_message = response.get("ai_message", {})

    if not ai_message:
        return None

    answer_text = ai_message.get("answer", "")

    raw_sources = ai_message.get("sources", [])
    sources = [
        {
            "title": s.get("title", ""),
            "link": s.get("link", ""),
            "source": s.get("source", "")
        }
        for s in raw_sources
    ]

    return {
        "answer": answer_text,
        "cited_sources": sources
    }
```

```python
def parse_baidu_health_box(response: dict) -> dict:
    """
    Extract the Baidu Health vertical box (jiankang.baidu.com).
    
    This is Baidu's dedicated health product — a structured health information
    card that appears at the top of results for medical queries. It is distinct
    from the general AI response and has not been studied in prior literature.
    
    May contain:
      - 'title': condition or drug name
      - 'description': short summary
      - 'sections': structured sections (symptoms, causes, treatment, etc.)
      - 'link': link to full Baidu Health page
    """
    health_box = response.get("health_results", {})

    if not health_box:
        return None

    return {
        "title": health_box.get("title", ""),
        "description": health_box.get("description", ""),
        "sections": health_box.get("sections", []),
        "link": health_box.get("link", "")
    }
```

```python
def parse_organic_results(response: dict, top_n: int = 5) -> list:
    """
    Extract top organic (non-AI, non-health-box) search results.
    
    Useful for understanding what non-AI sources Baidu surfaces,
    and for classifying source type (institutional, lay, government, etc.).
    
    Parameters
    ----------
    top_n : int
        How many organic results to capture (default 5).
    """
    organics = response.get("organic_results", [])
    return [
        {
            "title": r.get("title", ""),
            "link": r.get("link", ""),
            "snippet": r.get("snippet", ""),
            "source": r.get("source", "")
        }
        for r in organics[:top_n]
    ]
```

```python
def extract_related_searches(response: dict) -> list:
    """
    Extract Baidu's related searches (equivalent to Google's People Also Ask).
    
    Useful as seed queries for the next round of data collection.
    """
    return [
        item.get("query", "")
        for item in response.get("related_searches", [])
    ]
```

### Pipeline

For each query this pipeline captures:
1. **AI Message** — Baidu ERNIE's AI-generated answer (analog to Google AI Overview)
2. **Health Box** — Baidu Health vertical card (unique to Baidu; unstudied in literature)
3. **Organic Results** — Top 5 standard results (for source-type classification)
4. **Related Searches** — Baidu's equivalent of People Also Ask

Output columns mirror the Google notebook for easy cross-platform merging.

```python
def run_baidu_pipeline(queries: list, api_key: str) -> pd.DataFrame:
    """
    Run the full Baidu scraping pipeline over a list of (english_label, mandarin_query) tuples.
    
    Parameters
    ----------
    queries : list of (str, str)
        List of (english_label, mandarin_query) pairs.
    api_key : str
        SerpAPI key.

    Returns
    -------
    pd.DataFrame
        One row per query with all extracted fields.
    """
    rows = []
    total = len(queries)

    for i, (english_label, mandarin_query) in enumerate(queries):
        print(f"[{i+1}/{total}] EN: '{english_label}' | ZH: '{mandarin_query}'")

        try:
            # 1. Call Baidu via SerpAPI
            response = query_baidu(mandarin_query, api_key)

            # 2. Parse each result type
            ai_message   = parse_baidu_ai_message(response)
            health_box   = parse_baidu_health_box(response)
            organics     = parse_organic_results(response, top_n=5)
            related      = extract_related_searches(response)

            has_ai_message  = ai_message is not None
            has_health_box  = health_box is not None

            rows.append({
                # -- identifiers --
                "query_en":             english_label,
                "query_zh":             mandarin_query,

                # -- AI message (ERNIE) --
                "has_ai_message":       has_ai_message,
                "ai_answer":            ai_message["answer"]         if has_ai_message else None,
                "ai_cited_sources":     ai_message["cited_sources"]  if has_ai_message else None,

                # -- Baidu Health Box --
                "has_health_box":       has_health_box,
                "health_box_title":     health_box["title"]          if has_health_box else None,
                "health_box_desc":      health_box["description"]    if has_health_box else None,
                "health_box_sections":  health_box["sections"]       if has_health_box else None,
                "health_box_link":      health_box["link"]           if has_health_box else None,

                # -- Organic results --
                "organic_results":      organics,

                # -- Related searches (analog to PAA) --
                "related_searches":     related,
            })

        except Exception as e:
            print(f"   ERROR on '{mandarin_query}': {e}")
            rows.append({
                "query_en":             english_label,
                "query_zh":             mandarin_query,
                "has_ai_message":       False,
                "ai_answer":            None,
                "ai_cited_sources":     None,
                "has_health_box":       False,
                "health_box_title":     None,
                "health_box_desc":      None,
                "health_box_sections":  None,
                "health_box_link":      None,
                "organic_results":      None,
                "related_searches":     None,
            })

        time.sleep(1.5)

    df = pd.DataFrame(rows)

    print(f"\nTotal rows: {len(df)}")
    print(f"Queries with AI message:  {df['has_ai_message'].sum()}")
    print(f"Queries without AI message: {(~df['has_ai_message']).sum()}")
    print(f"Queries with Health Box:  {df['has_health_box'].sum()}")

    return df
```

```python
df = run_baidu_pipeline(queries, api_key)
df
```
